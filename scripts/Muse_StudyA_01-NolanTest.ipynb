```python
all_evokeds = []
#all_data = [] 

rej_thresh_uV = 100
rej_thresh = rej_thresh_uV*1e-6

#rejs = []
#drops = []

## loop over all subjects
for curr_subject in good_subjects:
    print("Current Subject #" + str(curr_subject))

    #Importing raw data
    raw = utils.load_data('../data/StudyA/', 
                          subject_nb = curr_subject, 
                          sfreq = 256., 
                          session_nb = [1,2],
                         );

    # Filtering data
    raw.filter(0.1, 15, method='iir', verbose= False);

    # Getting the events
    events = find_events(raw, min_duration=0);
    event_id = {'Positive': 1, 'Neutral': 2};
    print("Number of Events #" + str(len(events)))

    # Creating an Epoch object with standard parameters: Note: Picking only the ear-channels
    epochs = Epochs(raw, events=events, 
                    event_id=event_id, 
                    tmin=-0.1, 
                    tmax=0.8, 
                    baseline=(-0.1,0),
                    detrend = 1,
                    preload=True, 
                    verbose=False, 
                    picks=[0,3]); #these are only the ear channels
    #epochs.apply_baseline()
    
    
    #'''
    #Getting the automated rejection threshold for this file - otherwise same parameters as above
    print('------------ Autoreject: Get Threshold ------------')
    reject = get_rejection_threshold(epochs, decim=2)
    
    epochs = Epochs(raw, events=events, 
                    event_id=event_id, 
                    tmin=-0.1, 
                    tmax=0.8, 
                    baseline = (None, 0),
                    reject= {'eeg': rej_thresh}, 
                    detrend = 1,
                    preload=True, 
                    verbose=False, 
                    picks=[0,3]);#'''
    
    print('reject thresh: ', reject['eeg'])
    
    #rejs.append(reject['eeg'])
    print('------------ IMPORT DROP: ------------')
    epochs.plot_drop_log()
    epochs.drop_bad()#reject=reject)
    #epochs.drop_bad()

    
    # Performing Autorej.
    print('------------ Autoreject: Autocorrect ------------')
    ar = AutoReject( n_interpolate= [0], random_state=42, n_jobs=1, verbose=False );#verbose='tqdm')
    epochs_ar, reject_log = ar.fit_transform(epochs, return_log=True)
    

    
    # Plot Epochs & Epochs_AR
    epochs.plot_image(0, cmap='interactive', sigma=1.,)# vmin= -factor*reject['eeg'], vmax= factor*reject['eeg'])
    plt.show()
    
    epochs_ar.plot_image(0, cmap='interactive', sigma=1.,)# vmin= -factor*reject['eeg'], vmax= factor*reject['eeg'])
    plt.show()
    

    
    ###############
    '''import matplotlib
    plt.ion()
    %matplotlib qt
    #matplotlib.use('TkAgg')
    mne.viz.plot_epochs(epochs, picks=None, 
                        scalings=None, 
                        n_epochs=20, 
                        n_channels=2,
                        title=None, #events=events, event_colors={'Positive':'blue', 'Neutral':'red'}, 
                        show=True, block=True, 
                        decim='auto', 
                        noise_cov=None,
                        butterfly=True, 
                        show_scrollbars=False, 
                        epoch_colors=None,
                    );
    
    plt.ioff()
    plt.isinteractive()#'''
    
    #%matplotlib inline
    #print('------------ DONT KNOW ------------')

    #epochs.plot_drop_log();

    ###
    #events = find_events(raw, min_duration=0);
    events[:,2] = np.zeros(len(events[:,2]));
    #events[:,2]

    ###
    #rej_thresh_uV = 100
    #rej_thresh = rej_thresh_uV*1e-6
    #epochs2 = Epochs(raw, events=events, 
    #                    #event_id=event_id, 
    #                    tmin=-0.1, 
    #                    tmax=0.8, 
    #                    baseline=(-0.1,0),
    #                    reject= {'eeg': rej_thresh}, 
    #                    preload=True, 
    #                    verbose=False, 
    #                    picks=[0,3]);

    #epochs

    print('\n ----------------------------------------------')

    print('Raw has ' + str(len(events)) + ' events.')
    print('Epochs has ' + str(np.round(epochs['Neutral'].get_data().shape[0] + epochs['Positive'].get_data().shape[0], 2)) + ' trials.')
    print('Epochs_AR has ' + str(np.round(epochs_ar['Neutral'].get_data().shape[0] + epochs_ar['Positive'].get_data().shape[0], 2)) + ' trials.')
    
    print('\n ----------------------------------------------')

    
    ###
    #epochs.drop_bad(reject=reject)
    print('------------ AUTOREJ DROP LOG: ------------')
    epochs_ar.plot_drop_log();
    epochs_ar.drop_bad()
    #epochs.get_data().shape
    
    
    sample_drop = (1 - len(epochs_ar.events)/len(events)) * 100
    print('sample drop %: ', sample_drop)
    print('\n ----------------------------------------------')

    #drops.append(sample_drop)
    
    # Perform +/- Averaging to check signal quality
    new_array = epochs_ar.get_data().copy();

    for i in range(epochs_ar.get_data().shape[0]):
        if (i % 2) == 0:   #even number
            new_array[i,:,:] = new_array[i,:,:];
        elif (i % 2) != 0:   #uneven number 
            new_array[i,:,:] = -1 * new_array[i,:,:];
    epochs3  = mne.EpochsArray(new_array, epochs_ar.info, epochs_ar.events, epochs_ar.tmin, epochs_ar.event_id)
    #epochs3
    #curr = 3
    #plt.plot(epochs_ar.get_data()[curr,0,:]);
    #plt.plot(new_array[curr,0,:]); 
    #plt.show()

    mne.viz.plot_compare_evokeds([list(epochs_ar.iter_evoked()), 
                                  list(epochs3.iter_evoked())],
                                  colors = ['green','gray'],
                                  combine = 'mean',
                                  picks=[0,1])
    

    conditions = OrderedDict();
    conditions['1-Positive'] = [1];
    conditions['2-Neutral'] = [2];
    
    #curr_title = "Subject #" +str(curr_subject+1);

    if sample_drop < 100:
        if len(epochs_ar.events) > 60:
        
            evokeds = [epochs_ar[name].average() for name in ('Positive', 'Neutral')]

            mne.viz.plot_compare_evokeds(evokeds, 
                                         colors = ['red', 'blue'],
                                         picks = [0,1], 
                                         ylim  = dict(eeg=[-10, 10]),
                                         combine = 'mean'); 
            plt.show();

            all_evokeds.append(evokeds)

            #evokeds[0].data.shape  #pos
            #evokeds[1].data.shape  #neu
            #all_data.append(np.stack((   evokeds[0].data, evokeds[1].data))) 

            #plt.show() #'''

    print('\n----------------------------------------------------')
    print('---------------------------------------------------- \n')

#plt.bar(np.arange(len(drops)), drops); plt.show()

#all_data1 = np.asarray(all_data)
#all_data1.shape
```
